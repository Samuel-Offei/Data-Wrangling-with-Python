# 2. Handling Missing Values:!

a. Identify columns with missing values in the dataset and show the proportion of missing values in each column.!

```python
# Identify columns with missing values
missing_values = consumer_behavior.isnull().sum()

# Calculate the proportion of missing values in each column
missing_proportion = (missing_values / len(consumer_behavior)) * 100

# Create a DataFrame to display the results
missing_info = pd.DataFrame({
    'Mising Value': missing_values,
    'Missing Proportion (%)': missing_proportion
})

# Display the information
print(missing_info)
```

c. Apply appropriate techniques to handle missing values in these columns (e.g., imputation, removal, or other strategies).!

```python
# Impute Age with the mean
consumer_behavior['Age'].fillna(consumer_behavior['Age'].mean(), inplace=True)

# Impute Review Rating with the mode
consumer_behavior['Review Rating'].fillna(consumer_behavior['Review Rating'].mode()[0], inplace=True)

# Impute missing values in Purchase Amount (USD) with the mean
consumer_behavior['Purchase Amount (USD)'].fillna(consumer_behavior['Purchase Amount (USD)'].mean(), inplace=True)

# Remove rows with missing values in Item Purchased and Category
consumer_behavior.dropna(subset=['Item Purchased', 'Category', 'Location', 'Shipping Type', 'Color', 'Frequency of Purchases', 'Payment Method'], inplace=True)

# Impute Review Rating with the mode
consumer_behavior['Previous Purchases'].fillna(consumer_behavior['Previous Purchases'].mode()[0], inplace=True)

# Display the updated dataset with missing values handled
print("Updated Dataset with Missing Values Handled:")
consumer_behavior.head(10)
```

b. Discuss the importance of encoding categorical data for machine learning and analysis.

In machine learning and analysis, encoding categorical data is crucial because many algorithms require numerical input, and categorical variables, which represent different categories or labels, are often non-numeric. By encoding, we convert categories into a numerical format, ensuring that algorithms can process and understand the data correctly. This helps prevent misinterpretation of categorical variables and biases in the models.

c. Apply label encoding to all categorical columns and one-hot encoding all categorical column.

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
for column in consumer_behavior.select_dtypes(include=['object']).columns:
    consumer_behavior[column + '_label_encoded'] = label_encoder.fit_transform(consumer_behavior[column])
    
# Apply one-hot encoding to each categorical column
consumer_behavior_onehot = pd.get_dummies(consumer_behavior, columns=consumer_behavior.select_dtypes(include=['object']).columns, drop_first=True)
```
